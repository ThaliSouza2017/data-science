```
#importando bibliotecas
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```
#criando o dataset1
outs_withy = pd.read_csv('delta_withy_outs.csv')
outs_withy.head()
```

```
#criando o dataset2
outs_without = pd.read_csv('delta_without_outs.csv')
outs_without.head()
```

```
outs_withy.plot(x='delta', y='user_id')
plt.show()
```

```
outs_without.plot(x='delta', y='user_id')
plt.show()
```

```
outs_withy.hist('delta', color='purple')
plt.title('usuários COM registro de evento de ir à loja')
plt.ylabel('Número de usuários')
plt.xlabel('Distribuição de dias de permanência no app')
plt.rcParams['figure.figsize'] = (11 ,7)
#plt.legend(loc='best')
plt.show()

outs_without.hist('delta', color='orange')
plt.title('usuários SEM registro de evento de ir à loja')
plt.ylabel('Número de usuários')
plt.xlabel('Distribuição de dias de permanência no app')
plt.rcParams['figure.figsize'] = (11 ,7)
#plt.legend(loc='best')
plt.show()
```

```
outs_without_v2 = outs_without.query('delta >= 0')
outs_without_v2.head()
```
